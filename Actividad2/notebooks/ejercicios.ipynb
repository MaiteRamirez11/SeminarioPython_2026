### **2. Duración de una playlist**

Dada la siguiente playlist con canciones y sus duraciones en formato "m:ss":


````python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
````

 Calcule la duración total de la playlist en formato Xm Ys, y encuentre la canción más larga y la más corta.

**Nota:** Deberá convertir las duraciones de string a valores
numéricos para poder operar con ellas.

**Salida esperada:**
```
Duración total: 46m 53s
Canción más larga: "Stairway to Heaven" (8:02)
Canción más corta: "Imagine" (3:07)
```

````python
#Ejercicio 2
from src.funciones2 import convertir_a_segundos

playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]

#calcular duracion total en segundos de la playlist
total_segundos = sum(convertir_a_segundos(cancion["duration"]) for cancion in playlist)

#calcular minutos
minutos = total_segundos // 60

#calcular segundos
segundos = total_segundos % 60 

#funcion lambda asigna al parametro key la duración en segundos para encontrar el maximo o minimo segun corresponda
cancion_mas_larga = max(playlist, key=lambda c: convertir_a_segundos(c["duration"]))
cancion_mas_corta = min(playlist, key=lambda c: convertir_a_segundos(c["duration"]))

print(f"Duración total: {minutos}m {segundos}s")
print(f'Canción más larga: "{cancion_mas_larga["title"]}" ({cancion_mas_larga["duration"]})')
print(f'Canción más corta: "{cancion_mas_corta["title"]}" ({cancion_mas_corta["duration"]})')



````
